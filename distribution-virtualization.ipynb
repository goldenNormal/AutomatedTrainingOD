```python
import pandas as pd
import numpy as np
```

To obtain the results CSV file (see Readme.md), you need to run the corresponding experiment first.

```python
dataset = '18_Ionosphere.npz'
seed = 0
model = 'ae'


training = ['nai','optimal','en']
Path = [f'./{model}/{model}-{mode}-{dataset}-{seed}.csv' for mode in training]

optimal = np.array(pd.read_csv(Path[1])['AUC'])
naive = np.array(pd.read_csv(Path[0])['AUC'])
```

```python
linewidth = 1
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
auc = np.concatenate([naive,optimal])

label = np.concatenate([np.full(naive.shape,'Naive'),np.full(optimal.shape,'optimal')])

x = np.concatenate([np.full(naive.shape,1),np.full(naive.shape,1)])

df = pd.DataFrame({'auc':auc,'type':label,'vs':x})
# df = pd.DataFrame({'auc':optimal,'type':'optimal'})
sns.violinplot(data = df,x='vs',y='auc',hue='type', cut=0,split=True, inner='quartiles',linewidth=linewidth*1.7,scale='count',bw='scott')
```
